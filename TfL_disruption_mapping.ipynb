```python


# List of coordinates (lat, long) for the markers
coordinates = [
    (51.524672252099904, -0.1180309588427563)  # London House
]

f = folium.Figure(width=1000, height=500)
m = folium.Map(location= [51.524, -0.118], tiles="openstreetmap", zoom_start=16).add_to(f)

# Add markers to the map
for lat, lon in coordinates:
    folium.Marker(location=[lat, lon]).add_to(m)

# Display the map
m
```
